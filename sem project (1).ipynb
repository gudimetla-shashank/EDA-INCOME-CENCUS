```python
import numpy as np 
from scipy.stats import norm 
import matplotlib.pyplot as plt 

data = np.random.normal(adult['age']) 

# Fit a normal distribution to the data: 
# mean and standard deviation 
mu, std = norm.fit(data)  
  
# Plot the histogram. 
plt.hist(data, bins=25, density=True, alpha=0.6, color='y') 
  
# Plot the PDF. 
xmin, xmax = plt.xlim() 
x = np.linspace(xmin, xmax, 100) 
p = norm.pdf(x, mu, std) 
  
plt.plot(x, p, 'k', linewidth=2) 
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std) 
plt.title(title) 
  
plt.show()
```

# hypothesis testing

chi sqared test

```python
from scipy import stats

# Create a contingency table to represent the observed frequencies
contingency_table = pd.crosstab(adult['education'], adult['sex'])

# Calculate the expected frequencies based on the assumption of independence
expected_frequencies = stats.chi2_contingency(contingency_table)[1]

# Calculate the chi-squared statistic
chi2_statistic = stats.chi2_contingency(contingency_table)[0]

# Calculate the degrees of freedom
degrees_of_freedom = (contingency_table.shape[0] - 1) * (contingency_table.shape[1] - 1)

# Calculate the p-value
p_value = stats.chi2_contingency(contingency_table)[2]

# Print the results
print('Chi-squared statistic:', chi2_statistic)
print('Degrees of freedom:', degrees_of_freedom)
print('p-value:', p_value)

# Interpret the results
if p_value < 0.05:
    print('There is a significant difference in the distribution of rating levels across different sub-categories.')
else:
    print('There is no significant difference in the distribution of rating levels across different sub-categories.')
```

anova test

```python
from scipy.stats import f_oneway

age_categories = adult['age'].unique()

age_data = {category: adult['hours_per_week'][adult['age'] == category] for category in age_categories}

f_statistic, p_value = f_oneway(*age_data.values())

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant difference in mean hours per week among age categories.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant difference in mean hours per week among age categories.")
```
